```python
import os
import json
import pandas as pd

path = 'beer-database/'
dfs = []

for filename in os.listdir(path):
    if filename.endswith('.json'):
        filepath = os.path.join(path, filename)
        with open(filepath, 'r') as f:
            data = json.load(f)
        df = pd.json_normalize(data, 'data')
        dfs.append(df)

beer_df = pd.concat(dfs, ignore_index=True)

beer_df.head()
```
